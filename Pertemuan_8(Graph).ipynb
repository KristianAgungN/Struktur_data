Representasi Graf dengan Adjacency List


```python
# Membuat graf tak berarah menggunakan adjacency list
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

# Menampilkan graf
def print_graph(graf):
    for node in graf:
        print(f"{node} -> {graf[node]}")

print_graph(graph)
```

Traversal Graf menggunakan BFS

```python
from collections import deque

def bfs(graph, start):
    visited = set()
    queue = deque([start])


    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            print(vertex, end=" ")
            visited.add(vertex)
            queue.extend([neighbor for neighbor in graph[vertex] if neighbor not in visited])

# Contoh pemanggilan
bfs(graph, 'A')
```

Traversal Graf menggunakan DFS

```python
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()

    if start not in visited:
        print(start, end=" ")
        visited.add(start)
        for neighbor in graph[start]:
            dfs(graph, neighbor, visited)

# Contoh pemanggilan
dfs(graph, 'A')
```

```python
from collections import deque

# no1 (graph tak berarah minimal 5 simpul 7 sisi)
graph = {
    'A': ['B', 'C', 'D'],
    'B': ['A', 'C', 'E'],
    'C': ['A', 'B', 'D'],
    'D': ['A', 'C', 'E'],
    'E': ['B', 'D']
}

def print_graph(graph):
    print("Graph:")
    for node, neighbors in graph.items():
        print(f"  {node} : {neighbors}")

print_graph(graph)

# no2 (implementasi BFS dan DFS)
def bfs_print(graph, start):
    visited = set()
    queue = deque([start])
    visited.add(start)
    print(f"\nBFS dari '{start}':", end=" ")
    while queue:
        node = queue.popleft()
        print(node, end=" ")
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
    print()

def dfs_print(graph, start, visited=None):
    if visited is None:
        visited = set()
        print(f"\nDFS dari '{start}':", end=" ")
    visited.add(start)
    print(start, end=" ")
    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs_print(graph, neighbor, visited)

bfs_print(graph, 'A')
dfs_print(graph, 'A')
print()

# no3 (modifikasi fungsi BFS agar mengembalikan list urutan kunjungan)
def bfs(graph, start):
    visited = set()
    queue = deque([start])
    visited.add(start)
    order = []
    while queue:
        node = queue.popleft()
        order.append(node)
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
    return order

hasil_bfs = bfs(graph, 'A')
print(f"Urutan BFS dari 'A': {hasil_bfs}")

# no4 (buat fungsi find_path menggunakan DFS)
def find_path(graph, start, end, visited=None, path=None):
    if visited is None:
        visited = set()
        path = []
    visited.add(start)
    path = path + [start]
    if start == end:
        return path
    for neighbor in graph[start]:
        if neighbor not in visited:
            result = find_path(graph, neighbor, end, visited, path)
            if result:
                return result
    return None

print(f"\nJalur A -> E: {find_path(graph, 'A', 'E')}")
print(f"Jalur B -> D: {find_path(graph, 'B', 'D')}")

# no5 (tambahkan fungsi is_connected untuk mengecek apakah semua simpul saling terhubung)
def is_connected(graph):
    if not graph:
        return True
    start = next(iter(graph))
    visited = set()

    def dfs(node):
        visited.add(node)
        for neighbor in graph[node]:
            if neighbor not in visited:
                dfs(neighbor)

    dfs(start)
    return len(visited) == len(graph)

print(f"\nApakah graph terhubung? {is_connected(graph)}")

graph_terputus = {
    'A': ['B'],
    'B': ['A'],
    'C': ['D'],
    'D': ['C']
}
print(f"Graph terputus terhubung? {is_connected(graph_terputus)}")
```
