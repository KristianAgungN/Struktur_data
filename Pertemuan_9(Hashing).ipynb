Implementasi Hash Table Sederhana

```python
class SimpleHashTable:
  def __init__ (self, size):
    self.size = size
    self.table = [None] * size

  def hash_function (self, key):
    return hash (key) % self.size

  def insert (self, key, value):
    index = self.hash_function (key)
    self.table [index] = value

  def search (self, key) :
    index = self.hash_function (key)
    return self.table [index]

# Contoh penggunaan
ht = SimpleHashTable (10)
ht.insert("apel", 100)
ht.insert("pisang", 200)

print ("Nilai untuk 'apel':", ht.search("apel"))
print ("Nilai untuk 'pisang':", ht.search("pisang"))
```

Penanganan Collision Dengan Linear Probing

```python
class LinearProbingHashTable:
  def __init__(self, size):
    self.size = size
    self.table = [None] * size

  def hash_function (self, key):
    return hash (key) % self.size

  def insert (self, key, value):
    index = self.hash_function (key)
    original_index = index

    while self.table [index] is not None:
      if self.table [index] [0] == key:
        break
      index = (index + 1) % self.size
      if index == original_index:
        raise Exception ("Hash table penuh")

    self.table [index] = (key, value)

  def search (self, key) :
    index = self.hash_function (key)
    original_index = index

    while self.table [index] is not None:
      if self.table [index] [0] == key:
          return self.table [index] [1]
      index = (index + 1) % self.size
      if index == original_index:
        break
    return None

# Contoh penggunaan
ht = LinearProbingHashTable (5)
ht.insert("apel", 100)
ht.insert("pisang", 200)
ht.insert ("melon", 300)

print ("Nilai untuk 'apel':", ht.search("apel"))
print ("Nilai untuk 'melon': ", ht.search("melon"))
```

Penanganan Collision Dengan Chaining

```python
class ChainingHashTable:
  def __init__ (self, size):
    self.size = size
    self.table = [[] for _ in range (size)]

  def hash_function (self, key):
    return hash (key) % self.size

  def insert (self, key, value):
    index = self.hash_function (key)
    for i, (k, v) in enumerate (self.table [index]):
        if k == key:
            self.table [index] [i] (key, value)
            return
    self.table[index].append((key, value))

  def search(self, key):
      index = self.hash_function (key)
      for k, v in self.table[index]:
          if k == key:
              return v
      return None

# Contoh penggunaan
ht = ChainingHashTable (5)
ht.insert("apel", 100)
ht.insert("pisang", 200)
ht.insert("melon", 300)

print ("Nilai untuk 'apel':", ht.search("apel"))
print ("Nilai untuk 'melon': ", ht. search ("melon"))
```

Eksperimen Sederhana Linear Probing

```python
class LinearProbingHashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash_function(self, key):
        return hash(key) % self.size

    def insert(self, key, value):
        index = self.hash_function(key)
        original_index = index

        while self.table[index] is not None:
            if self.table[index][0] == key:
                break
            index = (index + 1) % self.size
            if index == original_index:
                raise Exception("Hash table penuh")

        self.table[index] = (key, value)

    def search(self, key):
        index = self.hash_function(key)
        original_index = index

        while self.table[index] is not None:
            if self.table[index][0] == key:
                return self.table[index][1]
            index = (index + 1) % self.size
            if index == original_index:
                break
        return None

# a. ukuran table 5
ht = LinearProbingHashTable(5)

# b. pasangan key-value
ht.insert("A", 10)
ht.insert("B", 20)
ht.insert("C", 30)
ht.insert("D", 40)
ht.insert("E", 50)

# c. cetak isi array
print("Isi self.table:", ht.table)

# d. apa yang terjadi jika dua key memiliki hasil hash yang sama?
#     data tidak akan tertimpa, dan linear probing akan mencari slot kosong.
```

Percobaan Dasar Chaining

```python
class ChainingHashTable:
    def __init__(self, size):
        self.size = size
        self.table = [[] for _ in range(size)]

    def hash_function(self, key):
        return hash(key) % self.size

    def insert(self, key, value):
        index = self.hash_function(key)
        for i, (k, v) in enumerate(self.table[index]):
            if k == key:
                self.table[index][i] = (key, value)
                return
        self.table[index].append((key, value))

    def search(self, key):
        index = self.hash_function(key)
        for k, v in self.table[index]:
            if k == key:
                return v
        return None

# a. ukuran tabel 3
ht = ChainingHashTable(3)

# b. key-value
ht.insert("apel", 100)
ht.insert("melon", 200)
ht.insert("lemon", 300)
ht.insert("pisang", 400)

# c. cetak isi tabel
print("Isi self.table:")
for i, slot in enumerate(ht.table):
    print(f"  Slot {i}:", slot)

# d. bagaimana metode chaining menangani collision?
#     collision ditangani dengan menambahkan data ke dalam list pada slot yang sama, sehingga data tidak tertimpa.
```
